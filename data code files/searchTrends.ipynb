```python
# Determine the size of each chunk, assuming the data is evenly distributed across the years
chunk_size = len(df['TD (degC)']) // 24

# Split the data into 24 chunks
chunks = [df['TD (degC)'][i*chunk_size:(i+1)*chunk_size] for i in range(24)]

# Calculate the average of each chunk
averages = [chunk.mean() for chunk in chunks]

averages
```
